### 加载为json数据

```python
data = json.loads(response.content.decode('utf-8'))
data
```

```python
html_elements = data['data']
```

```python
from lxml import etree
```

```python
# html = etree.HTML(html_elements)  # 初始化生成一个XPath解析对象
# result = etree.tostring(html, encoding='utf-8')  # 解析对象输出代码
html=etree.HTML(html_elements,etree.HTMLParser())
```

```python
for i in html:
    print(etree.tostring(i, encoding='utf-8').decode('utf-8'))
```

```python
# html = etree.tostring(html, encoding='utf-8').decode('utf-8')
trs = html.xpath('//tr')[1:]
```

```python
# 提取数据
days = {
    'date':[],
    'week':[],
    'max_temperature':[],
    'min_temperature':[],
    'day_status':[],
    'wind':[],
    'aqi':[],
    'aqi_status':[]
}




for tr in trs:
    data = tr.xpath('.//text()')
    # 消去垃圾值
    data = [x for x in data if x != '\n        ' and x != '    ']
    # 划分数据
    res_data = []
    for x in data:
        res_data += x.split()
    # print(res_data)
    
    
    # 添加数据
    days['date'].append(res_data[0])
    days['week'].append(res_data[1])
    days['max_temperature'].append(res_data[2])
    days['min_temperature'].append(res_data[3])
    days['day_status'].append(res_data[4])
    days['wind'].append(res_data[5])
    days['aqi'].append(res_data[6])
    days['aqi_status'].append(res_data[7])
```

```python
days
```

```python
import pandas  as pd
```

```python
df =  pd.DataFrame(days, index=days['date'])
```
